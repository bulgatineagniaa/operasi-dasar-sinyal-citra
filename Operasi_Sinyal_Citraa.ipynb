```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# --- 1. MEMPERSIAPKAN CITRA PERTAMA (Dari Bagian B.1) ---
# Menggunakan variabel img_rgb yang sudah sukses dimuat di cell sebelumnya
if 'img_rgb' in locals():
    img1 = img_rgb.copy()
else:
    # Antisipasi jika cell sebelumnya belum di-run
    img1 = np.ones((600, 800, 3), dtype=np.uint8) * 255

# --- 2. Membaca citra kedua (bola diputar)
img2 = cv2.imread('bola.jpg')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
img2 = cv2.rotate(img2, cv2.ROTATE_90_CLOCKWISE)

# --- 3. PROSES RESIZING (Penyelarasan Ukuran) ---
# Mengecek apakah ukuran citra 1 dan citra 2 berbeda
h1, w1, _ = img1.shape
h2, w2, _ = img2.shape

resizing_terjadi = False
if (h1 != h2) or (w1 != w2):
    # Menyamakan ukuran citra kedua mengikuti resolusi citra pertama
    img2_resized = cv2.resize(img2, (w1, h1), interpolation=cv2.INTER_LINEAR)
    resizing_terjadi = True
else:
    img2_resized = img2.copy()

# --- 4. OPERASI PENJUMLAHAN CITRA ---
# cv2.add mengimplementasikan teknik 'Clipping' secara otomatis (nilai max 255)
citra_hasil = cv2.add(img1, img2_resized)

# --- 5. MENAMPILKAN OUTPUT GRAFIK ---
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(img1)
plt.title(f'Citra 1 (Bola Pantai)\nUkuran: {w1}x{h1}')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(img2_resized)
plt.title(f'Citra 2 (Koin - Resized)\nUkuran Baru: {img2_resized.shape[1]}x{img2_resized.shape[0]}')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(citra_hasil)
plt.title('Citra Hasil Penjumlahan')
plt.axis('off')

plt.tight_layout()
plt.show()

# --- 6. PENJELASAN PROSES RESIZING ---
print("\n============ 4. PENJELASAN PROSES RESIZING ============")
if resizing_terjadi:
    print(f"-> Terjadi perbedaan resolusi awal:")
    print(f"   Citra 1 berukuran {w1}x{h1} piksel, sedangkan Citra 2 berukuran {w2}x{h2} piksel.")
    print(f"-> Tindakan: Dilakukan resizing pada Citra 2 menggunakan fungsi `cv2.resize()`")
    print(f"   dengan metode interpolasi bilinear (`INTER_LINEAR`) menjadi {w1}x{h1} piksel.")
    print(f"   Hasilnya, kedua matriks citra sekarang sinkron dan siap dijumlahkan elemen-per-elemen.")
else:
    print("-> Kedua citra sudah memiliki ukuran yang sama sejak awal, tidak diperlukan resizing.")
print("=======================================================")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1. Mengambil citra asli dari cell B.1
if 'img_rgb' in locals():
    img_asli = img_rgb.copy()
else:
    # Backup jika cell sebelumnya belum dijalankan
    img_asli = np.ones((600, 800, 3), dtype=np.uint8) * 255
    cv2.circle(img_asli, (400, 300), 100, (255, 0, 0), -1)

# Mendapatkan dimensi citra (Tinggi, Lebar)
h, w, _ = img_asli.shape

# 2. Menentukan kombinasi pergeseran sesuai soal
kombinasi = [
    {"di": 30, "dj": 0,  "label": "T1: di=30, dj=0 (Turun)"},
    {"di": 0,  "dj": 30, "label": "T2: di=0, dj=30 (Kanan)"},
    {"di": 30, "dj": 30, "label": "T3: di=30, dj=30 (Diagonal)"}
]

Citra_hasil_list = []

# 3. Melakukan Translasi menggunakan Matriks Affine 2x3
for shift in kombinasi:
    di = shift["di"]
    dj = shift["dj"]

    # Matriks Transformasi M = [[1, 0, dj], [0, 1, di]]
    M = np.float32([[1, 0, dj], [0, 1, di]])

    # Menggeser citra. Area kosong akan otomatis diisi warna hitam (0) oleh OpenCV
    img_shifted = cv2.warpAffine(img_asli, M, (w, h))
    Citra_hasil_list.append(img_shifted)

# 4. Menampilkan Grafik Perbandingan
plt.figure(figsize=(16, 10))

plt.subplot(2, 2, 1)
plt.imshow(img_asli)
plt.title('1. Citra Asli', fontsize=12)
plt.axis('on')

plt.subplot(2, 2, 2)
plt.imshow(Citra_hasil_list[0])
plt.title(f'2. {kombinasi[0]["label"]}', fontsize=12)
plt.axis('on')

plt.subplot(2, 2, 3)
plt.imshow(Citra_hasil_list[1])
plt.title(f'3. {kombinasi[1]["label"]}', fontsize=12)
plt.axis('on')

plt.subplot(2, 2, 4)
plt.imshow(Citra_hasil_list[2])
plt.title(f'4. {kombinasi[2]["label"]}', fontsize=12)
plt.axis('on')

plt.tight_layout()
plt.show()

# =========================================================================
# 5. KODE TAMBAHAN: OTOMATIS MENCETAK PENJELASAN AREA KOSONG SETELAH RUN
# =========================================================================
print("\n================= PENJELASAN AREA KOSONG SETELAH TRANSLASI =================")
print("1. Mengapa Area Kosong Muncul?:")
print("   Operasi translasi memindahkan seluruh piksel asli ke koordinat yang baru.")
print("   Ketika gambar digeser ke kanan (+dj) atau ke bawah (+di), baris dan kolom")
print("   piksel pada bagian paling atas dan kiri akan 'ditinggalkan' tanpa data.")
print("")
print("2. Karakteristik Fisik Area Kosong:")
print("   Karena ukuran kanvas gambar dipaksa tetap sama (800x600 piksel), OpenCV")
print("   secara otomatis mengisi koordinat yang kosong tersebut dengan nilai default 0.")
print("   Dalam domain warna, nilai 0 berarti ketiadaan cahaya, sehingga area kosong")
print("   tersebut secara visual tampak sebagai warna hitam pekat (Black Padding).")
print("")
print("3. Dampak Kehilangan Informasi (Clipping):")
print("   Munculnya area kosong hitam di satu sisi berbanding lurus dengan terbuangnya")
print("   data piksel di sisi sebaliknya. Piksel yang terdorong melebihi batas bingkai")
print("   kanvas (kanan dan bawah) akan dipotong dan hilang dari memori citra.")
print("============================================================================")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1. Mengambil citra asli dari cell B.1 (diubah ke Grayscale agar visualisasi histogram lebih mudah dipahami)
if 'img_rgb' in locals():
    img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)
else:
    # Backup jika belum di-run
    img_gray = (np.indices((300, 300))[0] / 300.0 * 255).astype(np.uint8)

# 2. Menentukan variasi nilai alpha yang diminta soal
alphas = [0.5, 1, 1.5, 2]

# 3. Membuat visualisasi subplot (4 baris x 2 kolom: Sisi kiri Citra, Sisi kanan Histogram)
plt.figure(figsize=(14, 18))

for i, alpha in enumerate(alphas):
    # Rumus operasi: I'(i,j) = alpha * I(i,j)
    # Menggunakan cv2.multiply untuk penanganan otomatis clipping/saturasi di angka 255
    img_amplified = cv2.multiply(img_gray, np.array([float(alpha)]))

    # ---- KOLOM 1: MENAMPILKAN CITRA ----
    plt.subplot(4, 2, 2*i + 1)
    plt.imshow(img_amplified, cmap='gray', vmin=0, vmax=255)
    if alpha == 1:
        plt.title(f'Citra Asli ($\\alpha = {alpha}$)', fontsize=12, fontweight='bold')
    else:
        plt.title(f'Citra Hasil Amplifikasi ($\\alpha = {alpha}$)', fontsize=12)
    plt.axis('off')

    # ---- KOLOM 2: MENAMPILKAN HISTOGRAM ----
    plt.subplot(4, 2, 2*i + 2)
    # Meratakan matriks menjadi 1D dengan .ravel() untuk membuat histogram
    plt.hist(img_amplified.ravel(), bins=256, range=(0, 256), color='purple', alpha=0.7)
    plt.title(f'Histogram Sinyal Piksel ($\\alpha = {alpha}$)', fontsize=11)
    plt.xlabel('Intensitas Piksel (0 - 255)')
    plt.ylabel('Frekuensi Kemunculan')
    plt.grid(True, linestyle=':', alpha=0.6)

plt.tight_layout()
plt.show()

# =========================================================================
# 4. KODE TAMBAHAN: OTOMATIS MENCETAK ANALISIS PERUBAHAN KE LAYAR COLAB
# =========================================================================
print("\n==================== 4. ANALISIS PERUBAHAN BRIGHTNESS DAN KONTRAS ====================")
print("• Pada alpha = 0.5 (Atenuasi):")
print("  Citra menjadi sangat redup/gelap. Pada grafik histogram, seluruh sebaran piksel")
print("  menyusut dan bergeser berkumpul ke arah kiri (mendekati 0/hitam). Kontras menurun.")
print("")
print("• Pada alpha = 1 (Asli):")
print("  Kondisi normal citra asli dengan rentang histogram sebaran ideal.")
print("")
print("• Pada alpha = 1.5 & 2.0 (Amplifikasi):")
print("  Citra mengalami peningkatan kecerahan (brightness) yang drastis sekaligus penguatan")
print("  kontras pada area abu-abu. Namun, terjadi penumpukan piksel dalam jumlah besar")
print("  di ujung paling kanan histogram (nilai 255), menandakan area terang yang jenuh (putih polos).")
print("======================================================================================")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Membuat Sinyal Input x[n] berupa sinyal diskrit sampel sederhana
n = np.arange(0, 10)
x_n = np.array([0, 1, 2, 3, 2, 1, 0, -1, -2, 0])

# 2. Menentukan 3 nilai alpha untuk pengujian
alphas = [0.5, 2, -1.5]

# 3. Membuat kanvas untuk plot perbandingan (3 baris x 2 kolom)
plt.figure(figsize=(12, 12))

print("=================== PROSES PERHITUNGAN SISTEM ===================")
print(f"Sinyal Input x[n] : {x_n}\n")

for i, alpha in enumerate(alphas):
    # Ruas Kiri: T(alpha * x) -> Input dikali alpha dulu, baru masuk sistem T(x)=2x
    input_dikali_alpha = alpha * x_n
    ruas_kiri = 2 * input_dikali_alpha

    # Ruas Kanan: alpha * T(x) -> Input masuk sistem T(x)=2x dulu, baru dikali alpha
    output_sistem_normal = 2 * x_n
    ruas_kanan = alpha * output_sistem_normal

    # Cetak data teks ke konsol
    print(f"• Pengujian Alpha = {alpha}:")
    print(f"  Hasil T(alpha * x) [Ruas Kiri]  : {ruas_kiri}")
    print(f"  Hasil alpha * T(x) [Ruas Kanan] : {ruas_kanan}\n")

    # ---- PLOT GRAFIK RUAS KIRI ----
    plt.subplot(3, 2, 2*i + 1)
    plt.stem(n, ruas_kiri, linefmt='b-', markerfmt='bo', basefmt='r-')
    plt.title(f'Ruas Kiri: $T(\\alpha \\cdot x[n])$ ($\\alpha = {alpha}$)')
    plt.ylabel('Amplitudo')
    plt.grid(True, linestyle=':', alpha=0.6)

    # ---- PLOT GRAFIK RUAS KANAN ----
    plt.subplot(3, 2, 2*i + 2)
    plt.stem(n, ruas_kanan, linefmt='g--', markerfmt='gx', basefmt='r-')
    plt.title(f'Ruas Kanan: $\\alpha \\cdot T(x[n])$ ($\\alpha = {alpha}$)')
    plt.ylabel('Amplitudo')
    plt.grid(True, linestyle=':', alpha=0.6)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Membuat Sinyal Input x1[n] dan x2[n] (Sinyal Diskrit Sampel)
n = np.arange(0, 10)
x1 = np.array([0, 1, 2, 3, 2, 1, 0, -1, -2, 0])
x2 = np.array([2, 2, 0, -1, -2, 0, 1, 2, 1, 0])

# 2. Perhitungan Matematika Sesuai Rumus Additivitas
# Ruas Kiri: T(x1 + x2) -> Sinyal dijumlahkan dulu, baru dimasukkan ke sistem (dikali 2)
sinyal_jumlah = x1 + x2
ruas_kiri = 2 * sinyal_jumlah

# Ruas Kanan: T(x1) + T(x2) -> Masing-masing dimasukkan ke sistem dulu, baru hasilnya dijumlahkan
tx1 = 2 * x1
tx2 = 2 * x2
ruas_kanan = tx1 + tx2

# Menghitung Selisih (Ruas Kiri - Ruas Kanan)
selisih = ruas_kiri - ruas_kanan

# 3. Membuat Plot Grafik (5 Baris Subplot sesuai Output Wajib)
plt.figure(figsize=(10, 14))

# Plot 1: Sinyal Input x1[n]
plt.subplot(5, 1, 1)
plt.stem(n, x1, linefmt='b-', markerfmt='bo', basefmt='r-')
plt.title('1. Plot Sinyal Input $x_1[n]$')
plt.ylabel('Amplitudo')
plt.grid(True, linestyle=':', alpha=0.6)

# Plot 2: Sinyal Input x2[n]
plt.subplot(5, 1, 2)
plt.stem(n, x2, linefmt='m-', markerfmt='mo', basefmt='r-')
plt.title('2. Plot Sinyal Input $x_2[n]$')
plt.ylabel('Amplitudo')
plt.grid(True, linestyle=':', alpha=0.6)

# Plot 3: T(x1 + x2)
plt.subplot(5, 1, 3)
plt.stem(n, ruas_kiri, linefmt='c-', markerfmt='co', basefmt='r-')
plt.title('3. Plot Ruas Kiri: $T(x_1 + x_2)$ (Dijumlahkan Dulu Baru Diproses)')
plt.ylabel('Amplitudo')
plt.grid(True, linestyle=':', alpha=0.6)

# Plot 4: T(x1) + T(x2)
plt.subplot(5, 1, 4)
plt.stem(n, ruas_kanan, linefmt='g--', markerfmt='gx', basefmt='r-')
plt.title('4. Plot Ruas Kanan: $T(x_1) + T(x_2)$ (Diproses Dulu Baru Dijumlahkan)')
plt.ylabel('Amplitudo')
plt.grid(True, linestyle=':', alpha=0.6)

# Plot 5: Selisih antara kedua hasil
plt.subplot(5, 1, 5)
plt.stem(n, selisih, linefmt='r-', markerfmt='ro', basefmt='r-')
plt.title('5. Plot Selisih (Ruas Kiri - Ruas Kanan)')
plt.xlabel('Sampel (n)')
plt.ylabel('Amplitudo')
plt.ylim(-2, 2) # Mengunci batas y agar terlihat jelas kalau nilainya 0
plt.grid(True, linestyle=':', alpha=0.6)

plt.tight_layout()
plt.show()

# Menampilkan data teks di bawah grafik
print("=================== DATA ANGKA UNTUK LAPORAN ===================")
print("1. Sinyal x1[n]             :", x1)
print("2. Sinyal x2[n]             :", x2)
print("3. Hasil T(x1 + x2)         :", ruas_kiri)
print("4. Hasil T(x1) + T(x2)      :", ruas_kanan)
print("5. Selisih Hasil            :", selisih)
print("================================================================")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sinyal input dasar
x = np.linspace(-10, 10, 100)
x1 = 3.0
x2 = 4.0
a = 2.0

# Fungsi Sistem
def T1(val): return 2 * val
def T2(val): return val ** 2

# Buat canvas grafik
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# ------------------------------------------------------------------
# [GRAFIK 1] UJI HOMOGENITAS T1 (LINIER)
# ------------------------------------------------------------------
axs[0, 0].plot(x, T1(x), label='Respon Sistem $T_1(x)=2x$', color='blue')
axs[0, 0].scatter([a * x1], [T1(a * x1)], color='red', s=100, zorder=5, label='$T_1(a \cdot x_1)$ (Sisi Kiri)')
axs[0, 0].scatter([a * x1], [a * T1(x1)], color='orange', marker='x', s=150, zorder=5, label='$a \cdot T_1(x_1)$ (Sisi Kanan)')
axs[0, 0].set_title("Uji Homogenitas $T_1(x) = 2x$\n(Titik Merah & 'X' Oranye Berhimpit = LOLOS)")
axs[0, 0].grid(True)
axs[0, 0].legend()

# ------------------------------------------------------------------
# [GRAFIK 2] UJI ADDITIVITAS T1 (LINIER)
# ------------------------------------------------------------------
axs[0, 1].plot(x, T1(x), label='Respon Sistem $T_1(x)=2x$', color='blue')
axs[0, 1].scatter([x1 + x2], [T1(x1 + x2)], color='purple', s=100, zorder=5, label='$T_1(x_1 + x_2)$ (Sisi Kiri)')
axs[0, 1].scatter([x1 + x2], [T1(x1) + T1(x2)], color='green', marker='x', s=150, zorder=5, label='$T_1(x_1) + T_1(x_2)$ (Sisi Kanan)')
axs[0, 1].set_title("Uji Additivitas $T_1(x) = 2x$\n(Titik Ungu & 'X' Hijau Berhimpit = LOLOS)")
axs[0, 1].grid(True)
axs[0, 1].legend()

# ------------------------------------------------------------------
# [GRAFIK 3] UJI HOMOGENITAS T2 (NON-LINIER)
# ------------------------------------------------------------------
axs[1, 0].plot(x, T2(x), label='Respon Sistem $T_2(x)=x^2$', color='red')
axs[1, 0].scatter([a * x1], [T2(a * x1)], color='red', s=100, zorder=5, label='$T_2(a \cdot x_1) = 36$')
axs[1, 0].scatter([a * x1], [a * T2(x1)], color='orange', marker='x', s=150, zorder=5, label='$a \cdot T_2(x_1) = 18$')
axs[1, 0].set_title("Uji Homogenitas $T_2(x) = x^2$\n(Titik Terpisah / Tidak Berhimpit = GAGAL)")
axs[1, 0].grid(True)
axs[1, 0].legend()

# ------------------------------------------------------------------
# [GRAFIK 4] UJI ADDITIVITAS T2 (NON-LINIER)
# ------------------------------------------------------------------
axs[1, 1].plot(x, T2(x), label='Respon Sistem $T_2(x)=x^2$', color='red')
axs[1, 1].scatter([x1 + x2], [T2(x1 + x2)], color='purple', s=100, zorder=5, label='$T_2(x_1 + x_2) = 49$')
axs[1, 1].scatter([x1 + x2], [T2(x1) + T2(x2)], color='green', marker='x', s=150, zorder=5, label='$T_2(x_1) + T_2(x_2) = 25$')
axs[1, 1].set_title("Uji Additivitas $T_2(x) = x^2$\n(Titik Terpisah / Tidak Berhimpit = GAGAL)")
axs[1, 1].grid(True)
axs[1, 1].legend()

plt.tight_layout()
plt.show()
```
